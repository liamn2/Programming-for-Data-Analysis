As we can see from the graphs for the four variables over time, there is no clear distibution to the data that we can extrapolate. The distribution of the data appears to be stochastic, that is, it is seemingly random and without a definable pattern. In this instance however, we need to be able to match the graph shapes to a standardised distribution in order to mimic the data results for our synthetic data creation. I will use the uniform data distribution as, from the graphs previously, this seems to standardise all of our graphs. 

```python
# import numpy
import numpy as np
import matplotlib.pyplot as plt
  
# Using uniform() method
gfg = np.random.uniform(1, 40, 100)
  
plt.hist(gfg, bins = 50, density = True)
plt.show()
```
